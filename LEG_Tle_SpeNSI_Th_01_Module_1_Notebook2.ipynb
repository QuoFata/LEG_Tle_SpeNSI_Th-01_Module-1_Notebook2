<code><mark>Implémentation de l'algorithme knn en programmation orientée objet (POO)</mark></code>

```
# -*- coding: utf-8 -*-
"""Algorithme ‘identificationIris'
- programmation orientée objet
- nombre de variables prédictrices : 4
- nombre de classes des données labellisées : 3
"""

# ===========================================================
#       IMPORT DE MODULES DE LA BIBLIOTHEQUE STANDARD
# ===========================================================

import csv
from math import sqrt

# ============================
#      CREATION DE CLASSES
# ============================

class Iris :
    
    # constructeur
    def __init__(self):
        # déclaration et initialisation des attributs
        self.__largeurSepale=0.0
        self.__longueurSepale=0.0
        self.__largeurPetale=0.0
        self.__longueurPetale=0.0
        self.__nomEspece=""
        self.__distance=0.0
    
    # définitions des méthodes
    
    # Accesseur (getter)
    def getLargeurSepale(self):
        return self.__largeurSepale
    
    def getLongueurSepale(self):
        return self.__longueurSepale
    
    def getLargeurPetale(self):
        return self.__largeurPetale
    
    def getLongueurPetale(self):
        return self.__longueurPetale
    
    def getNomEspece(self):
        return self.__nomEspece
    
    def getDistance(self):
        return self.__distance
    
    # Mutateurs (setters)
    def setLargeurSepale(self,val):
        self.__largeurSepale=val
    
    def setLongueurSepale(self,val):
        self.__longueurSepale=val
    
    def setLargeurPetale(self,val):
        self.__largeurPetale=val
    
    def setLongueurPetale(self,val):
        self.__longueurPetale=val
    
    def setNomEspece(self,val):
        self.__nomEspece=val
        
    def setDistance(self,irisEspeceInconnue):
        self.__distance=sqrt(pow(irisEspeceInconnue.getLongueurSepale()-self.__longueurSepale,2)+
                             pow(irisEspeceInconnue.getLargeurSepale()-self.__largeurSepale,2)+
                             pow(irisEspeceInconnue.getLongueurPetale()-self.__longueurPetale,2)+
                             pow(irisEspeceInconnue.getLargeurPetale()-self.__largeurPetale,2)
                             )
        
class PredictionEspece:
    
    # définition et initialisation des attributs
    def __init__(self):
        self.__nomEspece=""
        self.__nombreVoisins=0
    
    def getNomEspece(self):
        return self.__nomEspece
    def getNombreVoisins(self):
        return self.__nombreVoisins
    def setNomEspece(self,val):
        self.__nomEspece=val
    def incrementeNombreVoisins(self):
        self.__nombreVoisins=self.__nombreVoisins+1

# ===============================================
#   DÉCLARATION ET INITIALISATION DE VARIABLES
# ===============================================

# nom : jeuDonnesApprentissage (jeu des données d'apprentissage) | type : liste d'objets
jeuDonnesApprentissage=[]
with open('iris_complet.csv', newline='') as csvfile: 
    listecsv = csv.reader(csvfile, delimiter=',') 
    for row in listecsv:
        dataLab = Iris()
        dataLab.setLongueurSepale(float(row[0]))
        dataLab.setLargeurSepale(float(row[1]))
        dataLab.setLongueurPetale(float(row[2]))
        dataLab.setLargeurPetale(float(row[3]))
        dataLab.setNomEspece(row[4])
        jeuDonnesApprentissage.append(dataLab)
    csvfile.close()

# nom : irisEspeceInconnue (échantillon à identifier) | type = objet
irisEspeceInconnue=Iris()
irisEspeceInconnue.setLongueurSepale(float(input('Saisir la longueur du sépale de l\'iris inconnu : ')))
irisEspeceInconnue.setLargeurSepale(float(input('Saisir la largeur du sépale de l\'iris inconnu : ')))
irisEspeceInconnue.setLongueurPetale(float(input('Saisir la longueur du pétale de l\'iris inconnu : ')))
irisEspeceInconnue.setLargeurPetale(float(input('Saisir la largeur du pétale de l\'iris inconnu : ')))

# nom : k (nombre de voisins) | type : entier
k = int(input('Saisir le nombre de voisins voulu : '))     
    
# nom : nombreVoisinsParEspece (liste des noms des espèces utilisée pour la prédiction) | type : liste de d'objets
nombreVoisinsParEspece=[PredictionEspece(),PredictionEspece(),PredictionEspece()]
nombreVoisinsParEspece[0].setNomEspece("Iris-setosa")
nombreVoisinsParEspece[1].setNomEspece("Iris-versicolor")
nombreVoisinsParEspece[2].setNomEspece("Iris-virginica")

# ============================
#  PROGRAMME PRINCIPAL (MAIN)
# ============================

# calcul de la distance entre chaque éléments du jeu de données d'apprentissage et l'échantillon à identifier
for i in range(0,len(jeuDonnesApprentissage)):
    jeuDonnesApprentissage[i].setDistance(irisEspeceInconnue)
    
# tri du tableau d'objets "jeuDonnesApprentissage" par ordre croissant de l'attribut "distance"
jeuDonnesApprentissage.sort(key=lambda x:x.getDistance())

# comptage des labels des k premiers objets de la liste "jeuDonnesApprentissage"
for i in range(0,k):
    for j in range(len(nombreVoisinsParEspece)):
        if(jeuDonnesApprentissage[i].getNomEspece()==nombreVoisinsParEspece[j].getNomEspece()):
            nombreVoisinsParEspece[j].incrementeNombreVoisins()

# tri du tableau d'objets "nombreVoisinsParEspece" par ordre croissant de l'attribut "nombreVoisins"
nombreVoisinsParEspece.sort(key=lambda x:x.getNombreVoisins())

# Affichage de la prédiction : on affiche le nom de l'espèce du dernier élément de la liste "nombreVoisinsParEspece"
print( 'Prédiction : cet iris peut faire partie de l'espèce : ',nombreVoisinsParEspece[len(nombreVoisinsParEspece)-1].getNomEspece())
```
